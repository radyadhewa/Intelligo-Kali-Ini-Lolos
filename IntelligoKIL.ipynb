```python
from gensim.models import Word2Vec
```

```python
# Create a list of sentences for Word2Vec training
sentences = [str(merged_data['product_id'][i]) + ' ' + str(merged_data['seller_id'][i]) for i in range(len(merged_data))]
```

```python
# Tokenize the sentences
tokenized_sentences = [sentence.split() for sentence in sentences]
```

```python
# Train Word2Vec model
embedding_size = 100  # You can adjust the size of the embeddings
model = Word2Vec(tokenized_sentences, vector_size=embedding_size, window=5, min_count=1, sg=0)
```

```python
# Define a function to get the embedding vector for a given text
def get_embedding(text):
    tokens = text.split()
    vector = sum([model.wv[token] for token in tokens if token in model.wv])
    return vector
```

```python
# Apply the get_embedding function to 'product_id' and 'seller_id' columns
merged_data['product_id_embedding'] = merged_data['product_id'].apply(get_embedding)
merged_data['seller_id_embedding'] = merged_data['seller_id'].apply(get_embedding)
```
